```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the IPL dataset
df = pd.read_csv('matches.csv')
```

```python
# 1. Basic DataFrame Operations
# i) Show first n rows of the DataFrame
n = int(input("Enter n number of rows: "))  # You can change this value to show more or fewer rows
print(f'The first {n} rows are: ')
display(df.head(n))

# ii) Show last n rows of the DataFrame
print(f"\nLast {n} rows of the DataFrame:")
display(df.tail(n))

# iii) Stats, info, datatypes of each column of the DataFrame
print("\nDataFrame Info:")
df.info()

print("\nDescriptive Statistics:")
display(df.describe())

print("\nData Types of Each Column:")
print(df.dtypes)

# iv) Shape of the DataFrame
print(f"\nShape of the DataFrame: {df.shape}")

# v) Check null values and deal with them
print("\nNull Values in Each Column:")
null_values = df.isnull().sum()
print(null_values[null_values > 0])

# Fill null values with 0 (or you can choose another method)
df.fillna(0, inplace=True)

# vi) Check duplicates and drop them
duplicates = df.duplicated().sum()
print(f"\nNumber of duplicate rows: {duplicates}")

if duplicates > 0:
    df.drop_duplicates(inplace=True)
```

```python
# 14. Count of super overs played by teams (Countplot)
year = input("Enter year (all or 2008-2024): ")
team = input("Enter team name: ")
if year != 'all':
    data = df[(df['season'] == year) & (df['super_over'] == 1) & ((df['team1'] == team) | (df['team2'] == team))]
else:
    data = df[(df['super_over'] == 1) & ((df['team1'] == team) | (df['team2'] == team))]
plt.figure(figsize=(10, 5))
sns.countplot(x='season', data=data, order=sorted(data['season'].unique()))
plt.title(f'Super Over Matches Played by {team} in {year}')
plt.xlabel('Year')
plt.ylabel('Count')
plt.show()
```

```python
# 15. Year-wise average runs per innings by teams (Barplot)
year = input("Enter year (all or 2008-2024): ")
team = input("Enter team name: ")

# Approximate total runs per innings using target_runs column
df['approx_total_runs'] = df['target_runs'].fillna(0) + df['result_margin'].fillna(0)

if year != 'all':
    data = df[(df['season'] == year) & ((df['team1'] == team) | (df['team2'] == team))]
else:
    data = df[((df['team1'] == team) | (df['team2'] == team))]

data_grouped = data.groupby(['season', 'team1'])['approx_total_runs'].mean().reset_index()

plt.figure(figsize=(10, 5))
sns.barplot(x='season', y='approx_total_runs', data=data_grouped)
plt.title(f'Year-wise Average Runs per Innings by {team}')
plt.xlabel('Year')
plt.ylabel('Average Runs')
plt.xticks(rotation=45)
plt.show()
```

```python
# 16. Year-wise average runs per innings per venue (Barplot)
year = input("Enter year (all or 2008-2024): ")

if year != 'all':
    data = df[df['season'] == year].groupby(['venue'])['target_runs'].mean().reset_index()
    plt.figure(figsize=(14, 7))
    sns.barplot(x='venue', y='target_runs', data=data, palette='viridis')
    plt.title(f'Average Runs per Innings Per Venue in {year}')
else:
    data = df.groupby(['season', 'venue'])['target_runs'].mean().reset_index()
    plt.figure(figsize=(14, 7))
    sns.barplot(x='season', y='target_runs', hue='venue', data=data, dodge=True)
    plt.title(f'Average Runs per Innings Per Venue (All Years)')
    plt.legend(title="Venue", bbox_to_anchor=(1.05, 1), loc='upper left')

plt.xlabel("Venue" if year != 'all' else "Season")
plt.ylabel("Average Runs")
plt.xticks(rotation=45)
plt.show()
```

```python
# 17. Batting/Bowling pitch classification (Venue-wise)
venue = input("Enter venue: ")

if venue in df['venue'].unique():
    avg_runs = df[df['venue'] == venue]['result_margin'].mean()
    
    if avg_runs > 30:  # Threshold for classification
        print(f"{venue} is a Batting pitch.")
    else:
        print(f"{venue} is a Bowling pitch.")
else:
    print("Invalid venue entered.")
```
